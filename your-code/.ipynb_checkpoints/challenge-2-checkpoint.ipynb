## Coding Session I - 30 minutes

From the Fitbit data, we want to visually understand: 

1. **How the average number of steps change by month. Use the appropriate visualization to show the median steps by month?**
1. **Is Fitbitter more active on weekend or workdays?**

##### All plots must be in the same jupyter notebook cell.

Hints:

- Use `Months_encoded` and `Week_or Weekend` columns
- Use `matplolib.pyplot` object oriented API
- Set your size figure to `12,4`
- Explore `plt.sca`
- Explore `plt.xticks`
- Save your figures

```python
months_steps = fitbit.groupby('Months_encoded', as_index=False).agg({'Steps': 'mean'})
x1 = months_steps.Months_encoded
y1 = months_steps.Steps
week_steps = fitbit.groupby('Work_or_Weekend', as_index=False).agg({'Steps': 'mean'})
x2 = week_steps.Work_or_Weekend
y2 = week_steps.Steps
```

```python
# Driver 1 enter codes here
fig, (ax1, ax2) = plt.subplots(2,1)
fig = plt.figure(figsize=(12,4))
ax1.plot(x1,y1, marker='o', ls='--', animated=True)
ax2.bar(x2,y2)
ax1.set_xlabel('Months')
ax1.set_ylabel('Steps')
ax1.set_xticks(x1)
ax2.set_ylabel('Steps')
ax2.set_xticks(x2)
for index in range(len(x1)):
  ax1.text(x1[index], y1[index], int(y1[index]), size=10)
for index in range(len(x2)):
  ax2.text(x2[index], y2[index], int(y2[index]), size=12)
```

```python
'''
The average number of steps is lower during the first two months of the year and then it goes up during the hotter months.
When autumn starts, the average of steps starts a declinng trend

Fitbitter is a little bit more active during the workdays (almost 500 steps more than on the weekend)
```

## Coding Session II - 30 minutes

**Write a loop to plot 3 scatter plots of the following features:**

1. **`Minutes Lightly Active`  vs `Steps`**
1. **`Minutes Very Active` vs `Steps`**
1. **`Minutes Sedentary` vs `Steps`**

```python
# Driver 2 enter codes here
x = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
y = ['Steps']

fitbit_loop = fitbit[['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary', 'Steps']]

fig, ax = plt.subplots(3,1, figsize=(40, 60))
for i in range(len(x)):
    ax[i].scatter(fitbit[fitbit_loop.keys()[i]], fitbit[fitbit_loop.keys()[-1]])
    ax[i].set_xlabel(fitbit_loop.keys()[i], fontsize=36)
    ax[i].tick_params(axis='both', labelsize=24)
    ax[i].set_ylabel(fitbit_loop.keys()[-1], fontsize=36)
    i+=1
```
